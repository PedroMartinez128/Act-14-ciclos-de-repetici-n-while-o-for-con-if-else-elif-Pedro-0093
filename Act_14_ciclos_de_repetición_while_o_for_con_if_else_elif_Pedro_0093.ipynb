##### Programa que calcule el Mínimo Común Múltiplo (MCM) de dos números enteros positivos usando un bucle de búsqueda o la relación con el MCD.
#### Prompt
#### Lenguaje de programación:Python
#### Puedes generar un Programa que calcule el Mínimo Común Múltiplo (MCM) de dos números enteros positivos usando un bucle de búsqueda o la relación con el MCD.

* OJO no funciones, no programación orientada a objetos, no listas, no diccionarios, no try.


```python
# 1. ENTRADA DE DATOS
a = int(input("Introduce el primer número entero positivo: "))
b = int(input("Introduce el segundo número entero positivo: "))

# Guardamos los valores originales para el cálculo final
num1 = a
num2 = b

# 2. PROCESO (Bucle para encontrar el MCD)
while b != 0:
    residuo = a % b
    a = b
    b = residuo

# Al salir del bucle, 'a' es el MCD
mcd = a

# Calculamos el MCM usando la relación matemática con el MCD
mcm = (num1 * num2) // mcd

# 3. SALIDA DE RESULTADOS
print("-----------------------------------------")
print("Resultados para los números", num1, "y", num2, ":")
print("El Máximo Común Divisor (MCD) es:", mcd)
print("El Mínimo Común Múltiplo (MCM) es:", mcm)
print("-----------------------------------------")
```
